```python
from scripts.util import get_day_range_from_string

n_dates_all, n_dates_found_all, n_articles_all, n_articles_dec_found_all = 0, 0, 0, 0
for key, page in parsed_pages.items():
    n_dates, n_dates_found = len(get_day_range_from_string(page.title)), len(page.sections)
    n_articles, n_articles_dec_found = 0, 0
    for d_s in page.sections:
        n_art, n_decs = len(d_s.article_sections), \
                          sum(int(len(a_s.decisions) > 0) for a_s in d_s.article_sections)
        n_articles, n_articles_dec_found = n_art + n_articles, n_articles_dec_found + n_decs
    n_dates_all, n_dates_found_all, n_articles_all, n_articles_dec_found_all = n_dates_all + n_dates, n_dates_found_all + n_dates_found, n_articles_all + n_articles, n_articles_dec_found_all + n_articles_dec_found
    print(f'{key}: {n_dates_found:02d}/{n_dates:02d} Days ({n_dates_found / n_dates * 100 if n_dates > 0 else 0:.2f}%); {n_articles_dec_found}/{n_articles} Articles ({n_articles_dec_found / n_articles * 100 if n_articles > 0 else 0:.2f}%)')
print(f'All: {n_dates_found_all:02d}/{n_dates_all:02d} Days ({n_dates_found_all / n_dates_all * 100 if n_dates_all > 0 else 0:.2f}%); {n_articles_dec_found_all}/{n_articles_all} Articles ({n_articles_dec_found_all / n_articles_all * 100 if n_articles_all > 0 else 0:.2f}%)')
```
